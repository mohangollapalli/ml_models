```python
import numpy as np
from sklearn import datasets 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from regression import LinearRegression
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=12)
indices = range(X.shape[0])
```

```python
X_train, X_test, y_train, y_test, train_idx, test_idx = train_test_split(X, y, indices, test_size=0.20, random_state=12)
```

```python
model = LinearRegression(lr=0.01)
model.fit(X_train, y_train)
test_preds = model.predict(X_test)
```

```python
# Eval. metrics 

def eval_metrics(actual_y, pred_y): 
    mse = (1/len(actual_y)) * np.sum((actual_y - pred_y)**2)
    rmse = np.sqrt(mse)
    return mse, rmse
```

```python
mse, rmse = eval_metrics(y_test, test_preds)
mse, rmse 
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error
mean_squared_error(y_test, test_preds), root_mean_squared_error(y_test, test_preds)
```

```python
fig, ax = plt.subplots()
ax.scatter(X_train, y_train, color='g', s=10, marker='o')
ax.scatter(X_test, y_test, color='b', s=10, marker='*')
ax.plot(X_test, test_preds, color='red')
ax.set_xlabel("X")
ax.set_ylabel("y")
plt.show()
```
